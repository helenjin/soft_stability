```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilons = torch.tensor([1e-1, 1e-2, 1e-3])
deltas = torch.tensor([1e-1, 1e-2, 1e-3])
Ns = torch.arange(10, 1000)
```

```python
fs = 12
plt.clf()
fig, ax = plt.subplots(figsize=(4,2))

for d in deltas:
    ax.plot(Ns, (torch.log(2/d) / (2*Ns)).sqrt(), label=f"$\delta =${d:.2e}")

# ax.legend(title="Fail Prob. $\delta$", loc="upper right", title_fontsize=fs, fontsize=fs)
ax.legend(loc="upper right", title_fontsize=fs, fontsize=fs)
ax.set_ylabel("Confidence Intv ($\epsilon$)", fontsize=fs)
ax.set_xlabel("Number of Samples", fontsize=fs)
ax.set_yscale("log")
ax.tick_params(axis="both", labelsize=fs-2)
plt.savefig("../figures/epsilon_vs_N_for_fixed_delta.pdf", bbox_inches="tight")
```

```python
plt.clf()
fig, ax = plt.subplots(figsize=(4,2))

for eps in epsilons:
    ax.plot(Ns, 2 / (2 * Ns * (eps ** 2)), label=f"$\epsilon = ${eps:.1e}")

# ax.legend(title="Conf Intv. $\epsilon$", loc="lower left", title_fontsize=fs, fontsize=fs)
ax.legend(loc="upper right", title_fontsize=fs, fontsize=fs)
ax.set_ylabel("Failure Prob. ($\delta$)", fontsize=fs)
ax.set_xlabel("Number of Samples", fontsize=fs)
ax.set_yscale("log")
ax.tick_params(axis="both", labelsize=fs-2)
plt.savefig("../figures/delta_vs_N_for_fixed_epsilon.pdf", bbox_inches="tight")
```
